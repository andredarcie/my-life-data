```python
import matplotlib.pyplot as plt
import pandas as pd
values = df['watched'].value_counts()
print(values)
# Supondo que a variável `values` foi gerada com `df['watched'].value_counts()`
# Ajuste para transformar em DataFrame com colunas renomeadas
values = values.reset_index()
values.columns = ['Ano', 'Quantidade']  # Renomear colunas para clareza

# Ordena os dados pelo ano
values = values.sort_values(by='Ano')

# Configuração do gráfico
fig, ax = plt.subplots(figsize=(12, 7))
ax = values.plot.bar(x='Ano', y='Quantidade', rot=45, ax=ax, color='skyblue', edgecolor='black')

# Melhorias visuais
ax.set_title("Filmes Assistidos por Ano", fontsize=16, weight='bold')
ax.set_xlabel("Ano", fontsize=14, weight='bold')
ax.set_ylabel("Número de Filmes", fontsize=14, weight='bold')
ax.legend(["Filmes Assistidos"], fontsize=12)
ax.bar_label(ax.containers[0], label_type='edge', fontsize=12, padding=3)

# Estilo da grade
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibição do gráfico
plt.show()
```
